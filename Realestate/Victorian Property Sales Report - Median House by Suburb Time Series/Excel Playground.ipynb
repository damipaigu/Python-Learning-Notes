```python
data = pd.read_excel('suburb_house_2017.xls', header=None, skiprows=[0, 1, 2],
                     names=['locality', 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 'prelim_18', 'change_16-17', 'change_07-17', 'growth_PA'])
```

```python
data = data.set_index('locality')
data.columns.names = ['median']
# data.index.names = ['locality']
data.head()
```

```python
data.insert(13, 'change_17-18', ((data['prelim_18'] - data[2017])/data[2017]*100).round(2))
```

```python
# 房价年平均增长率最高的地区
data.loc[data.growth_PA.idxmax()]
```

```python
# 房价年增长率最低的地区
data.loc[data.growth_PA.idxmin()]
```

```python
data.sort_values(by='prelim_18', ascending=False)
```

```python
data[data['prelim_18'] < 1200000].sort_values(by='prelim_18')
```

```python
# check which suburbs drops price and which increase during the last year
increase_bool = data['change_17-18'] > 0
decrease_bool = data['change_17-18'] <= 0
```

```python
increase_data = data[increase_bool]
decrease_data = data[decrease_bool]
```
